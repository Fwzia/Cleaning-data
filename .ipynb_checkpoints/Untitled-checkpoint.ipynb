```python
import pandas as pd
```

```python
books = pd.read_csv('books.csv')
```

```python
books.columns
```

```python
books = books[['id','isbn' ,'best_book_id', 'original_title', 'books_count', 'average_rating']]
```

```python
books.head()
```

```python
books.shape
```

```python
books.isna().sum()
```

```python
books = books.dropna(axis=0)
```

```python
books.shape
```

```python
#Merge 2 datasets in deffrent excel sheet
book_tags = pd.read_csv('book_tags.csv')
book_tags = book_tags[['goodreads_book_id', 'tag_id']]
books = books.merge(book_tags, left_on="best_book_id", right_on="goodreads_book_id")
books = books.drop('goodreads_book_id', axis = 1)
```

```python
books.head()
```

```python
import itertools
unique_id = list(set(itertools.chain(books.best_book_id)))
print(unique_id)
len(unique_id)
```

```python
unique_books_ids = {}
for i in range(len(unique_id)):
    unique_books_ids[unique_id[i]] = i
print(unique_books_ids)
```
